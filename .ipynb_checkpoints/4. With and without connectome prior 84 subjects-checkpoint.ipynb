### compute connectome transition probas

```python
transition_probabilities = []
for x in tqdm_notebook(X):
    x = np.log1p(x)
    x = (1/x)
    x[np.isinf(x)]=0
    g = nx.from_numpy_array(x)
    path_len = dict(nx.all_pairs_bellman_ford_path_length(g))
    spl = np.zeros((68,68))
    for i in range(68):
        for j in range(68):
            spl[i,j] = path_len[i][j]
    spl_proba = np.exp(-spl)
    np.fill_diagonal(spl_proba, 0)
    spl_proba = spl_proba / spl_proba.sum(axis=1).reshape(-1,1)
    transition_probabilities.append(spl_proba)
transition_probabilities = np.array(transition_probabilities)
```

$$
\frac{1}{\log(w_{ij})}
$$

```python
P = transition_probabilities.copy()
#P_aver = np.tile(transition_probabilities.mean(axis=0), (84, 1, 1))
```

```python
X.shape
```

```python
x = X.mean(axis=0)
x = np.log1p(x)
x = (1/x)
x[np.isinf(x)]=0
g = nx.from_numpy_array(x)
path_len = dict(nx.all_pairs_bellman_ford_path_length(g))
spl = np.zeros((68,68))
for i in range(68):
    for j in range(68):
        spl[i,j] = path_len[i][j]
spl_proba = np.exp(-spl)
np.fill_diagonal(spl_proba, 0)
spl_proba = spl_proba / spl_proba.sum(axis=1).reshape(-1,1)
```

```python
spl_proba.shape
```

```python
P_aver = np.tile(spl_proba, (84, 1, 1))
```

```python
all_orders_prior = []
for r in range(20):
    experiment = MCMC(LogLikelihood(prior_order_proba=P), n_burn_iter=20_000, random_state=r)
    best_order = experiment.run_experiment(thick, y, use_prior=True)
    
    max_loglike = np.argsort(experiment.mcmc_loglike_sequence)[::-1]
    all_orders_prior.append(np.array(experiment.all_orders)[max_loglike[:10]])
all_orders_prior = np.concatenate(all_orders_prior)
```

```python
all_orders_aver_prior = []
for r in range(20):
    experiment = MCMC(LogLikelihood(prior_order_proba=P_aver), n_burn_iter=20_000, random_state=r)
    best_order = experiment.run_experiment(thick, y, use_prior=True)
    
    max_loglike = np.argsort(experiment.mcmc_loglike_sequence)[::-1]
    all_orders_aver_prior.append(np.array(experiment.all_orders)[max_loglike[:10]])
all_orders_aver_prior = np.concatenate(all_orders_aver_prior)
```

# Kendall tau patient order agreement with time

```python
def kendall(a):
    n = a.shape[0]
    if n==1:
        return 1
    elif n==2:
        if a[0] <= a[1]:
            return 1
        else:
            return -1
    elif n==3:
        score = 0
        if a[0] <= a[1]:
            score+=1
        else:
            score-=1
            
        if a[0] <= a[2]:
            score+=1
        else:
            score-=1
            
        if a[1] <= a[2]:
            score+=1
        else:
            score-=1
        
        return score / 3
```
